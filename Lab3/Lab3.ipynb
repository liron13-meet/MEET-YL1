```python
import csv

f=open('cig.csv','rb')
r=csv.reader(f)
x=[]
y=[]
for row in r:
    row=row[0].replace("\t"," ")
    row=row.split()
    row=row[1:]
    row=map(float,row)
    x.append(row[0])
    y.append(row[1])

f.close()

from pylab import *

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title("nicotine and tar content")
ax.set_xlabel('tar content ')
ax.set_ylabel('nicotine content ')

scatter(x,y)

savefig("cig_scatter_labels.png")
```
